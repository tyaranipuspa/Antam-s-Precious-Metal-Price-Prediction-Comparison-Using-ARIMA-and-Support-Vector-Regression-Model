```
#IMPORT LIBRARY
import pandas as pd
import numpy as np
from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Panel, Tabs
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
from bokeh.models.tools import HoverTool
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT
```

```
#INPUT DATASET
stock = pd.read_csv('stock_market.csv')
stock
```

```
#RENAME COLUMNS
stock.rename(columns = {'Adj Close': 'Adj_Close'}, inplace=True)
stock.head()
```

```
#MEMBAGI DATASET BERDASARKAN NAMA INDEKS PASAR SAHAM
a = stock[(stock.values.ravel() == "HANG SENG").reshape(stock.shape).any(1)]
b = stock[(stock.values.ravel() == "NASDAQ").reshape(stock.shape).any(1)]
c = stock[(stock.values.ravel() == "NIKKEI").reshape(stock.shape).any(1)]

#MENGUBAH FORMAT DATA
a["Date"] = pd.to_datetime(a["Date"], format='%Y/%m/%d')
b["Date"] = pd.to_datetime(b["Date"], format='%Y/%m/%d')
c["Date"] = pd.to_datetime(c["Date"], format='%Y/%m/%d')

#MEMBUAT SOURCE DATA UNTUK GRAFIK
sourceHS = ColumnDataSource(a)
sourceNQ = ColumnDataSource(b)
sourceNK = ColumnDataSource(c)
```

```
#NUMBER 1

#Memanggil fungsi untuk keluar grafik di notebook yang sama
output_notebook()

#membuat figure
no1 = figure(x_axis_type='datetime', x_axis_label='Date', y_axis_label='Adj Close')
no1.line(x="Date", y="Adj_Close", source = sourceHS, line_width=2, legend_label="Hang Seng", color='deepskyblue')
no1.line(x="Date", y="Adj_Close", source = sourceNQ, line_width=2, legend_label="Nasdaq", color='lime')
no1.line(x="Date", y="Adj_Close", source = sourceNK, line_width=2, legend_label="Nikkei", color='crimson')
no1.legend.location = 'top_right'

#membuat hover
hover1 = HoverTool()
hover1.tooltips = [('Name','@Name'),('Adj Close','@Adj_Close')]
no1.add_tools(hover1)

#menampilkan grafik
show(no1)
```

```
#NUMBER 2

#Memanggil fungsi untuk keluar grafik di notebook yang sama
output_notebook()

#Membuat grafik tab Adj Close
fig1 = figure(x_axis_type='datetime', x_axis_label='Date', y_axis_label='Adj Close')
fig1.line(x="Date", y="Adj_Close", source = sourceHS, line_width=2, legend_label="Hang Seng", color='deepskyblue')
fig1.line(x="Date", y="Adj_Close", source = sourceNQ, line_width=2, legend_label="Nasdaq", color='lime')
fig1.line(x="Date", y="Adj_Close", source = sourceNK, line_width=2, legend_label="Nikkei", color='crimson')
fig1.legend.location = 'top_right'

#Membuat hover Adj Close
hfig1 = HoverTool()
hfig1.tooltips = [('Name','@Name'),('Adj Close','@Adj_Close')]
fig1.add_tools(hfig1)

#Membuat tab Adj Close
tab1 = Panel(child=fig1, title="Adj Close")

#Membuat grafik tab Volume
fig2 = figure(x_axis_type='datetime', x_axis_label='Date', y_axis_label='Volume')
fig2.line(x="Date", y="Volume", source = sourceHS, line_width=2, legend_label="Hang Seng", color='deepskyblue')
fig2.line(x="Date", y="Volume", source = sourceNQ, line_width=2, legend_label="Nasdaq", color='lime')
fig2.line(x="Date", y="Volume", source = sourceNK, line_width=2, legend_label="Nikkei", color='crimson')
fig2.legend.location = 'top_left'

#Membuat hover Volume
hfig2 = HoverTool()
hfig2.tooltips = [('Name','@Name'),('Volume','@Volume')]
fig2.add_tools(hfig2)

#Membuat tab Volume
tab2 = Panel(child=fig2, title="Volume")

#Membuat grafik tab Day_Perc_Change
fig3 = figure(x_axis_type='datetime', x_axis_label='Date', y_axis_label='Day_Perc_Change')
fig3.line(x="Date", y="Day_Perc_Change", source = sourceHS, line_width=2, legend_label="Hang Seng", color='deepskyblue')
fig3.line(x="Date", y="Day_Perc_Change", source = sourceNQ, line_width=2, legend_label="Nasdaq", color='lime')
fig3.line(x="Date", y="Day_Perc_Change", source = sourceNK, line_width=2, legend_label="Nikkei", color='crimson')
fig3.legend.location = 'top_right'

#Membuat hover Day_Perc_Change
hfig3 = HoverTool()
hfig3.tooltips = [('Name','@Name'),('Day_Perc_Change','@Day_Perc_Change')]
fig3.add_tools(hfig3)

#Membuat tab Day_Perc_Change
tab3 = Panel(child=fig3, title="Day_Perc_Change")

#Membuat grafik tab keseluruhan
no2 = Tabs(tabs= [ tab1, tab2, tab3 ])

#Menampilkan grafik
show(no2)
```
